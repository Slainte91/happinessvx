```python
csv_file = "wages_bycountry.csv"
wages_df = pd.read_csv(csv_file)
```

```python
wages_df.head(5)


lines = wages_df['Gross Average Monthly Wages by Country and Year']
```

```python
lines=lines[1:]
```

```python
def country_wage(row):
    i = 32
    country1=""
    while(i<len(row)):
        country1 = country1+row[i]
        i=i+1
        if (row[i] == ";"):
            break;
    country=country1[:-1] 
    country

    wage = ""
    i=i+1;
    while(i<len(row)):
        wage = wage+row[i]
        i=i+1


    return country,wage
```

```python
country_list=[]
wage_list=[]
for line in lines:
    country,wage = country_wage(line)
    country_list.append(country)
    wage_list.append(wage)
    
 
    
    
```

```python
wage_df = pd.DataFrame(list(zip(country_list,wage_list)), columns=['Country' , 'Wage'])

wage_df



```

```python
new_dataset = pd.merge(happiness_df,wage_df , how ='inner')
new_dataset
```

```python
new_dataset=new_dataset[new_dataset.Wage !='..']
new_dataset
```

```python
new_dataset.to_csv("new_dataset.csv" , index = False , header = True) 
```
